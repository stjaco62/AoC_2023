```python
# Puzzle 1

def get_data(filename):
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            line = line.split(",")
    return line

def calc_hash(code):
    value = 0
    for char in code:
        value += ord(char)
        value *= 17
        value = value % 256
    return value

def initialize_boxes():
    l_boxes = []
    for i in range(256):
        l_lenses = []
        l_boxes.append(l_lenses)
    return l_boxes

def compile_code(l_code):
    new_code = []
    for code in l_code:
        if "=" in code:
            lbl, lgth = code.split("=")
            lgth = int(lgth)
            new_code.append(("=", lbl, lgth))
        elif "-" in code:
            lbl = code[:-1]
            new_code.append(("-", lbl, 0))
    return new_code

def exec_equal(lbl, lgth):
    global l_boxes
    box_no = calc_hash(lbl)
    l_lenses = l_boxes[box_no]
    is_in_box = False
    for lens in l_lenses:
        if lbl == lens[0]:
            lens[1] = lgth
            is_in_box = True
    if not is_in_box:
        l_lenses.append([lbl, lgth])

def exec_minus(lbl):
    global l_boxes
    box_no = calc_hash(lbl)
    l_lenses = l_boxes[box_no]
    is_in_box = False
    for lens in l_lenses:
        if lbl == lens[0]:
            lgth = lens[1]
            is_in_box = True
    if is_in_box:
        l_lenses.remove([lbl, lgth])
    
def run_code(l_code):
    for code in l_code:
        op, lbl, lgth = code
        if op == "=":
            exec_equal(lbl, lgth)
        elif op == "-":
            exec_minus(lbl)
        else:
            print("Error: Unknown Operand")
            
def calc_focusing_power():
    global l_boxes
    focusing_power = 0
    for box_num in range(len(l_boxes)):
        for slot_num in range(len(l_boxes[box_num])):
            focusing_power += (box_num + 1) * (slot_num + 1) * (l_boxes[box_num][slot_num][1])
    return focusing_power
                


# Main Program, Puzzle 1
l_code = get_data("Puzzle.txt")

total = 0
for code in l_code:
    total += calc_hash(code)
print("Puzzle 1: The total is:", total)

# Puzzle 2 starts here
l_boxes = initialize_boxes()
l_code = compile_code(l_code)
run_code(l_code)

focusing_power = calc_focusing_power()
print("Puzzle 2: The focusing power is:", focusing_power)
```
