```python
# Puzzle 1

def get_data(filename):
    space = []
    with open(filename, "r") as file:
        for line in file:
            row = []
            line = line.strip()
            for char in line:
                row.append(char)
            space.append(row)
    return space

def print_space(space):
    for row in space:
        for col in row:
            print(col, end = "")
        print()
        
def is_row_empty(row):
    for char in row:
        if char == "#":
            return False
    return True

def travers_matrix(m):
    tr_m = []
    for col in range(len(m[0])):
        new_row = []
        for row in m:
            new_row.append(row[col])
        tr_m.append(new_row)
    return tr_m
        
def expand_space(space, expansion_factor):
    new_space = []
    expansion_row = []
    for i in range(len(space[0])):
        expansion_row.append(expansion_factor)
    for row in space:
        if is_row_empty(row):
            new_space.append(expansion_row)
        else:
            new_space.append(row)
    
    space = travers_matrix(new_space)
    new_space = []
    expansion_row = []
    for i in range(len(space[0])):
        expansion_row.append(expansion_factor)
    for row in space:
        if is_row_empty(row):
            new_space.append(expansion_row)
        else:
            new_space.append(row)
        
    space = travers_matrix(new_space)
    return space

def put_ones(space, expansion_factor):
    new_space = []
    for row in space:
        new_row = []
        for x in row:
            if x == expansion_factor:
                new_row.append(expansion_factor)
            else:
                new_row.append(1)
        new_space.append(new_row)
    return new_space

def collect_galaxies(space):
    l_gal = []
    for r in range(len(space)):
        for c in range(len(space[0])):
            if space[r][c] == "#":
                l_gal.append((r,c))
    return l_gal

def calc_distance(space, g1, g2):
    r1, c1 = g1
    r2, c2 = g2
    rmin = min(r1, r2)
    rmax = max(r1, r2)
    cmin = min(c1, c2)
    cmax = max(c1, c2)
    total = 0
    for r in range(rmin, rmax):
        total += space[r][cmin]
    for c in range(cmin, cmax):
        total += space[rmin][c]
    return total
    

def calc_total_distance(space, l_gal):
    tot_dist = 0
    for g1 in l_gal:
        for g2 in l_gal:
            tot_dist += calc_distance(space, g1, g2)
    return tot_dist // 2



# Main Program: Set expansion_factor to 2 to solve Puzzle1 and to 1000000 to solve Puzzle2
space = get_data("Puzzle.txt")

expansion_factor = 1000000

l_gal = collect_galaxies(space)
space = expand_space(space, expansion_factor)

space = put_ones(space, expansion_factor)

total = calc_total_distance(space, l_gal)
print("Total Distance: ", total)
```
