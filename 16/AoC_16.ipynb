```python
# Puzzle1

def get_data(filename):
    l_tiles = []
    with open(filename, "r") as file:
        for line in file:
            row = []
            line = line.strip()
            for char in line:
                row.append(char)
            l_tiles.append(row)
    return l_tiles

def energize(row, col, dir):
    global l_status
    l_status[row][col].append(dir)

def beam(row, col, dir):
    #print(row, col, dir)
    global l_tiles
    global l_status
    global l_beams
    if (row < 0) or (row >= len(l_tiles)) or (col < 0) or (col >= len(l_tiles[0])):
        return
    if dir in l_status[row][col]:
        return
    energize(row, col, dir)
    if l_tiles[row][col] == ".":
        if dir == "R":
            beam(row, col + 1, "R")
        elif dir == "L":
            beam(row, col -1, "L")
        elif dir == "U":
            beam(row - 1, col, "U")
        elif dir == "D":
            beam(row + 1, col, "D")
    elif l_tiles[row][col] == "/":
        if dir == "R":
            beam(row - 1, col, "U")
        elif dir == "L":
            beam(row + 1, col, "D")
        elif dir == "U":
            beam(row, col + 1, "R")
        elif dir == "D":
            beam(row, col - 1, "L")
    elif l_tiles[row][col] == "\\":
        if dir == "R":
            beam(row + 1, col, "D")
        elif dir == "L":
            beam(row - 1, col, "U")
        elif dir == "U":
            beam(row, col - 1, "L")
        elif dir == "D":
            beam(row, col + 1, "R")
    elif l_tiles[row][col] == "-":
        if dir == "R":
            beam(row, col + 1, "R")
        elif dir == "L":
            beam(row, col - 1, "L")
        elif (dir == "U") or (dir == "D"):
            beam(row, col - 1, "L")
            l_beams.append((row, col + 1, "R"))
    elif l_tiles[row][col] == "|":
        if dir == "U":
            beam(row - 1, col, "U")
        elif dir == "D":
            beam(row + 1, col, "D")
        elif (dir == "R") or (dir == "L"):
            beam(row + 1, col, "D")
            l_beams.append((row - 1, col, "U"))
        
def create_status_tiles(l_tiles):
    l_status = []
    for row in range(len(l_tiles)):
        st_row = []
        for col in range(len(l_tiles[0])):
            st_row.append([])
        l_status.append(st_row)
    return l_status
            
def count_energy_tiles(l_status):
    total = 0
    for row in range(len(l_status)):
        for col in range(len(l_status[0])):
            if l_status[row][col] != []:
                total += 1
    return total

def reset_status_tiles():
    global l_status
    for row in range(len(l_status)):
        for col in range(len(l_status[0])):
            l_status[row][col] = []
                   
def try_starting_position(row, col, dir):
    global l_status
    global l_tiles
    global l_beams
    reset_status_tiles()
    l_beams = [(row, col, dir)]
    while l_beams != []:
        b = l_beams.pop()
        row, col, dir = b
        beam(row, col, dir)
    #print_status_tiles()
                           
    total = count_energy_tiles(l_status) 
    return total
    
def print_status_tiles():
    global l_status
    for row in range(len(l_status)):
        for col in range(len(l_status[0])):
            if l_status[row][col] == []:
                print(".", end = "")
            else: 
                print("#", end = "")
        print()


# Main Program
l_beams = []
l_tiles = get_data("Puzzle.txt")
l_status = create_status_tiles(l_tiles)
max_total = 0
for row in range(len(l_tiles)):
    total = try_starting_position(row, 0, "R")
    if total > max_total:
        max_total = total
    total = try_starting_position(row, len(l_tiles[0]) - 1, "L")
    if total > max_total:
        max_total = total
for col in range(len(l_tiles[0])):
    total = try_starting_position(0, col, "D")
    if total > max_total:
        max_total = total
    total = try_starting_position(len(l_tiles) - 1, col, "U")
    if total > max_total:
        max_total = total
    
print(max_total)
```
