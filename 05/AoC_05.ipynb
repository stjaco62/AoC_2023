```python
#puzzle 1

def read_data_p1(filename):
    maps = []
    seeds = {}

    file = open(filename, "r")
    inp = file.read()
    inp = inp.split("\n\n")
    
    # Get seeds from first line   
    data = inp[0].strip()
    data = data.split(": ")[1].split(" ")
    for i in data:
        seeds[int(i)] = int(i)
    del inp[0]
    
    # Get maps 
    for data in inp:
        mp = []
        data = data.split(":\n")[1]
        data = data.split("\n")
        for t in data:
            t = t.split(" ")
            t = (int(t[0]), int(t[1]), int(t[2]))
            mp.append(t)
        maps.append(mp)
    
    return seeds, maps
    
def map_x_y(seeds, mp):
    mapped = {}
    for seed in seeds:
        mapped[seed] = False
    for m in mp:
        dest, src, rng = m
        for seed in seeds:
            if (seeds[seed] >= src) and (seeds[seed] < src + rng) and not(mapped[seed]):
                seeds[seed] = seeds[seed] + dest - src
                mapped[seed] = True
    return seeds
    
def find_min(seeds):
    l = []
    for seed in seeds:
        l.append(seeds[seed])
    print(min(l))

    
seeds, maps = read_data_p1("test.txt")

print(seeds)

for mp in maps:
    seeds = map_x_y(seeds, mp)
    print("Mapping", seeds)

find_min(seeds)
```

```python
# puzzle 2

def read_data_p2(filename):
    maps = []
    seeds = []

    file = open(filename, "r")
    inp = file.read()
    inp = inp.split("\n\n")
    
    # Get seeds from first line   
    data = inp[0].strip()
    data = data.split(": ")[1].split(" ")
    while len(data) > 0:
        strt = int(data[0])
        rng = int(data[1])
        seed = (strt, strt + rng - 1)
        seeds.append(seed)
        del data[0:2]
    del inp[0]
    
    # Get maps 
    for data in inp:
        mp = []
        data = data.split(":\n")[1]
        data = data.split("\n")
        for t in data:
            t = t.split(" ")
            t = (int(t[0]), int(t[1]), int(t[2]))
            mp.append(t)
        maps.append(mp)
    
    return seeds, maps

def shift(x, s):
    return x + s

def split_int(min1, max1, min2, max2, s):
    # Depending on overlap with mapping-intervals, split seed-intervals
    shifted_seed = []
    unshifted_seed = []
    if max1 < min2:
        unshifted_seed.append((min1, max1))
    if max1 >= min2 and max1 <= max2 and min1 < min2:
        unshifted_seed.append((min1, min2 - 1))
        shifted_seed.append((shift(min2, s), shift(max1, s)))
    if max1 <= max2 and min1 >= min2:
        shifted_seed.append((shift(min1, s), shift(max1, s)))
    if min1 > max2:
        unshifted_seed.append((min1, max1))
    if max1 > max2 and min1 >= min2 and min1 <= max2:
        unshifted_seed.append((max2 + 1, max1))
        shifted_seed.append((shift(min1, s), shift(max2, s)))
    if min1 < min2 and max1 > max2:
        unshifted_seed.append((min1, min2 - 1))
        unshifted_seed.append((max2 + 1, max1))
        shifted_seed.append((shift(min2, s), shift(max2, s)))
    return shifted_seed, unshifted_seed
    
def map_x_y(seeds, mp):
    shifted_seeds = []
    for m in mp:
        dest, src, rng = m
        min2 = src
        max2 = src + rng - 1
        shft = dest - src
        unshifted_seeds = []
        for seed in seeds:
            min1, max1 = seed
            shifted_s, unshifted_s = split_int(min1, max1, min2, max2, shft)
            #print("TPL", tpl)
            shifted_seeds += shifted_s
            unshifted_seeds += unshifted_s
        seeds = unshifted_seeds
        #print("Seeds", seeds)
    return shifted_seeds + unshifted_seeds
        
def find_min(seeds):
    l = []
    for seed in seeds:
        l.append(seed[0])
    print(min(l))
    
    
    
seeds, maps = read_data_p2("puzzle.txt")

print(seeds)
#print(maps)

for mp in maps:
    seeds = map_x_y(seeds, mp)
    print(mp)
    print()
    print("Mapping", seeds)
    
find_min(seeds)

        
```
