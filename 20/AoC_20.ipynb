```python
# Puzzle 1
# In Flip-Flops: status off -> False, status on --> True
# Pulses: low pulse --> False, high pulse --> True

def get_data(filename):
    d_modules = {}
    with open(filename, "r") as file:
        for line in file:
            line = line.strip()
            input, output = line.split(" -> ")
            output = output.split(", ")
            if input == "broadcaster":
                spec = {"type" : "broadcaster", "output" : output}
                d_modules[input] = spec
                for module in output:
                    if not(module in d_modules):
                        d_modules[module] = {"type" : "output", "output" : []}
            elif input[0] == "%":
                spec = {"type" : "flip-flop", "status" : False, "output" : output}
                d_modules[input[1:]] = spec
                for module in output:
                    if not(module in d_modules):
                        d_modules[module] = {"type" : "output", "output" : []}
            elif input[0] == "&":
                spec = {"type" : "conjunction", "status" : {}, "output" : output}
                d_modules[input[1:]] = spec
                for module in output:
                    if not(module in d_modules):
                        d_modules[module] = {"type" : "output", "output" : []}
        return (d_modules)

def print_modules():
    for name, module in d_modules.items():
        print(name, module)

def initialize_conjunction():
    global d_modules
    for name, module in d_modules.items():
        for outp in module["output"]:
            if d_modules[outp]["type"] == "conjunction":
                d_modules[outp]["status"][name] = False
    return d_modules

def exec_broadcaster(pulse):
    global d_modules
    global l_signals

    for outp in d_modules["broadcaster"]["output"]:
        l_signals.append(("broadcaster", outp, pulse))
               
def exec_flipflop(module, pulse):
    global d_modules
    global l_signals
    
    if pulse: # Pulse is high
        pass
    else:     # Pulse is low
        d_modules[module]["status"] = not(d_modules[module]["status"])
        for outp in d_modules[module]["output"]:
            l_signals.append((module, outp, d_modules[module]["status"]))
            
def exec_conjunction(inp_module, module, pulse):
    global d_modules
    global l_signals
    
    d_modules[module]["status"][inp_module] = pulse
    # Check, if all statuses are True (high), only then total_status is True
    total_status = True
    for stat in d_modules[module]["status"]:
        if not d_modules[module]["status"][stat]:
            total_status = False
    if total_status:
        for outp in d_modules[module]["output"]:
            l_signals.append((module, outp, False))
    else: 
        for outp in d_modules[module]["output"]:
            l_signals.append((module, outp, True))
            
def exec_output(pulse):
    pass  
                            
def exec_next_command(signal):
    global d_modules
    global total_high 
    global total_low
    
    inp, out, pulse = signal
    if pulse: # High pulse
        total_high += 1
    else:
        total_low += 1
        
    # Puzzle 2: Check if machine is "rx" and pulse is low
    if inp == "qk" and pulse:
        print("Inp", inp, "Outp", out, "Pulse", pulse, "No of button pressed:", button_press)
        
    if out == "broadcaster":
        exec_broadcaster(pulse)
    elif d_modules[out]["type"] == "flip-flop":
        exec_flipflop(out, pulse)
    elif d_modules[out]["type"] == "conjunction":
        exec_conjunction(inp, out, pulse)
    elif d_modules[out]["type"] == "output":
        sig = exec_output(pulse)
    else:
        print("Error: Unknown Type", signal)
    
    
# Main Program Puzzle 1
d_modules = get_data("Puzzle.txt")
d_modules = initialize_conjunction()
print_modules()
l_signals = []
l_records = []

steps = 0
total_high = 0
total_low = 0
button_press = 0

for i in range(10000):
    button_press += 1
    l_signals.append(("button", "broadcaster", False))
    while (l_signals != []):
        next_signal = l_signals[0]
        l_records.append(next_signal)
        exec_next_command(next_signal)
        del l_signals[0]
        steps += 1

print("Steps: ", steps, "\nTotal High:", total_high, "\nTotal Low:", total_low, "\nResult: ", total_high * total_low)
```
