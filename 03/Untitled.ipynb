```python
# puzzle 1

def get_input():
    engine = []
    with open("test.txt", "r") as file:
        for line in file:
            line = line.strip()
            line = "..." + line + "..."
            row = []
            for char in line:
                row.append(char)
            engine.append(row)
        row = []
        for i in range(len(engine[0])):
            row.append(".")
        for i in range(3):
            engine.append(row)
            engine.insert(0, row)
        return engine
    
def print_engine(engine):
    for row in engine:
        for char in row:
            print(char, end="")
        print()
        
def is_adjacent(engine, row_ind, col_ind):
    non_symbol_list = [str(x) for x in range(10)]
    non_symbol_list.append(".")
    for x in range(-1, 2):
        for y in range(-1, 2):
            if not(engine[row_ind + x][col_ind + y] in non_symbol_list):
                return True
    return False           
        
def search_numbers(engine):
    l_num = []
    for row_ind in range(len(engine)):
        num = 0
        num_adj = False
        for col_ind in range(len(engine[row_ind])):
            if engine[row_ind][col_ind].isdigit():
                num = 10 * num + int(engine[row_ind][col_ind])
                num_adj = num_adj or is_adjacent(engine, row_ind, col_ind)
            else:
                if not(num == 0) and num_adj:
                    l_num.append(num)
                num = 0
                num_adj = False
    return l_num

def make_part(engine, row_ind, col_ind):
    non_symbol_list = [str(x) for x in range(10)]
    non_symbol_list.append(".")
    part = []
    for i in range(-1, 2):
        new_row = engine[row_ind + i]
        new_row = new_row[col_ind - 3 : col_ind + 4]
        new_row.append(".")
        new_row.insert(0, ".")
        part.append(new_row)
    new_row = []
    for i in range(len(part[0])):
        new_row.append(".")
    part.append(new_row)
    part.insert(0, new_row)
    for i in range(len(part)):
        for j in range(len(part[0])):
            if not(part[i][j] in non_symbol_list):
                part[i][j] = "."
    part[2][4] = "*"           
    #print_engine(part)
    return part
    
def search_gears(engine):
    total_power = 0
    for row_ind in range(len(engine)):
        for col_ind in range(len(engine[0])):
            if engine[row_ind][col_ind] == "*":
                part = make_part(engine, row_ind, col_ind)
                l = search_numbers(part)
                if len(l) == 2:
                    power = l[0] * l[1]
                    #print(power)
                    total_power += power
    return total_power
                    
                

engine = get_input()
#print_engine(engine)
print(sum(search_numbers(engine)))
print(search_gears(engine))


    
    
    
    
```
