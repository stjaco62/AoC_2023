```python
# Puzzle1, first solution. Does not work for the second part due to limited time and memory
def get_data(filename):
    garden = []
    with open(filename, "r") as file:
        for line in file:
            row = []
            line = line.strip()
            for char in line:
                if not char == "S":
                    row.append(char)
                else:
                    row.append("O")
            garden.append(row)
    return garden

def print_garden():
    global garden
    for row in garden:
        for col in row:
            print(col, end = "")
        print()
    print()

def create_empty_garden(garden):
    new_garden = []
    for row in garden:
        new_row = []
        for col in row:
            if col == "#":
                new_row.append("#")
            elif col == "S":
                new_row.append("S")
            else:
                new_row.append(".")
        new_garden.append(new_row)
    return new_garden
    
def go_one_step(garden):
    new_garden = create_empty_garden(garden)
    for r, row in enumerate(garden):
        for c, col in enumerate(row):
            if garden[r][c] in ["S", "O"]:
                if (0 <= r - 1) and (garden[r - 1][c] != "#"):
                    new_garden[r - 1][c] = "O"
                if (r + 1 < len(garden)) and (garden[r + 1][c] != "#"):
                    new_garden[r + 1][c] = "O"
                if (0 <= c - 1) and (garden[r][c - 1] != "#"):
                    new_garden[r][c - 1] = "O"
                if (c + 1 < len(row)) and (garden[r][c + 1] != "#"):
                    new_garden[r][c + 1] = "O"
    return new_garden

def count_garden_plots(garden):
    count = 0
    for row in garden:
        for col in row:
            if col == "O":
                count += 1
    return count
                                     
                    
# Main Program
steps = 64
garden = get_data("Puzzle.txt")
garden[65][65] = "."
garden[0][0] = "O" 




while steps > 0:
    steps -= 1
    garden = go_one_step(garden)
    #print_garden()
    
total = count_garden_plots(garden)
print("Total: ", total)
```

```python
import numpy as np

points = [(i, calc(65 + i * 131)) for i in range(3)]

def evaluate_quadratic_equation(points, x):
    # Fit a quadratic polynomial (degree=2) through the points
    coefficients = np.polyfit(*zip(*points), 2)

    # Evaluate the quadratic equation at the given x value
    result = np.polyval(coefficients, x)
    return round(result)

evaluate_quadratic_equation(points, 202300)  
```
